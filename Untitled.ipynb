```python
def TrueError(ActualValue, EstimatedValue):
    return ActualValue - EstimatedValue

def RelativeError(ActualValue, EstimatedValue):
    TrueError = ActualValue - EstimatedValue
    return TrueError / ActualValue

def ApproximateError(PreviousEstimation, CurrentEstimation):
    return CurrentEstimation - PreviousEstimation
    
def ApproximateRelativeError(PreviousEstimation, CurrentEstimation):
    ApproximateError = CurrentEstimation - PreviousEstimation
    return ApproximateError / CurrentEstimation
```

```python
#For the Taylor series to Compute an estimate of non-polynomial function we use cos(x) centered at pi

import math
pi = math.pi

def TaylorCoefficient(n):
    return (-1)**(n+1)/(math.factorial(2n))

def TaylotPolynomialTerms(n, input):
    return TaylorCoefficient(n)*((input-pi)**2n)

def TalylorPolynomial(n, input):
    sum = 0
    for i in range(0, n+1):
        sum = sum + TaylorPolynomialTerms(i, input)
        return sum
```

```python
n = [i for i in range(0,21)]
ActualValue = [math.cos(pi+.1) for i in n]
EstimatedValue = [TaylorPolynomial(x, pi + .1) for x in n]
True_Error = [TrueError(ActualValue[i], EstimatedValue[i]) for i in n]
Relative_Error = [RelativeError(ActualValue[i],EstimatedValue[i]) for i in n]
Approximate_Error = [ApproximateError(EstimatedValue[i], EstimatedValue[i+1]) for i in range (0,20)]
Approximate_Relative_Error = [ApproximateRelativeError(EstimatedValue[i], EstimatedValue[i+1]) for i in range (0,20)]
```
